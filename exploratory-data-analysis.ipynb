# Z- score method
**Z score is also called as Standard score about a data point how much about mean 
2.Z score detects outlier based threshold value
            Z =|x - mean|/std**

```python
outliers=[]
def detect_outlier(data1):
    threshold = 2.0
    mean = np.mean(data1)
    std = np.std(data1)
    for x in data1:
        z_score =np.abs((x-mean)/std)
        if np.ceil(z_score)> threshold:
            outliers.append(x)
    return outliers
outliers =detect_outlier(data1)
outliers
```
